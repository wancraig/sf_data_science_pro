```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sber_data = pd.read_csv('data/sber_data.csv')
sber_data.head()
```

```python
display(sber_data.tail())
```

```python
# Сколько районов Москвы и Московской области представлено в данных?
unique_areas = sber_data['sub_area'].nunique()
print(f"Количество районов Москвы и Московской области в данных: {unique_areas}")
```

```python
# Чему равна максимальная цена квартир (price_doc)? Введите это число полностью, без округлений.
max_price = sber_data['price_doc'].max()
print(f"Максимальная цена квартир: {max_price}")
```

```python
# Проверим, влияет ли уровень экологической обстановки в районе на цену квартиры. Постройте коробчатую диаграмму цен
# на квартиры (price_doc) в зависимости от уровня экологической обстановки в районе (ecology).
# Какой уровень ценится на рынке меньше всего? Введите название этой категории без кавычек.
# Построение коробчатой диаграммы
plt.figure(figsize=(10, 6))
sns.boxplot(x='ecology', y='price_doc', data=sber_data)
plt.xlabel('Уровень экологической обстановки')
plt.ylabel('Цена квартиры')
plt.title('Зависимость цены квартиры от уровня экологии')
plt.xticks(rotation=45)  # Поворот подписей по оси X для удобства чтения

plt.tight_layout()
plt.show()
```

```python
# Постройте диаграмму рассеяния, которая покажет, как цена на квартиру (price_doc) связана с расстоянием до центра Москвы
# (kremlin_km). Выберите все верные утверждения.
# Построение диаграммы рассеяния
plt.figure(figsize=(8, 6))
sns.scatterplot(x='kremlin_km', y='price_doc', data=sber_data)
plt.xlabel('Расстояние до центра Москвы, км')
plt.ylabel('Цена квартиры')
plt.title('Связь цены квартиры с расстоянием до центра Москвы')

plt.tight_layout()
plt.show()
```

```python
# ищем пропуски
display(sber_data.isnull().tail())
```

```python
# получаем список столбцов с пропусками
cols_null_percent = sber_data.isnull().mean() * 100
cols_with_null = cols_null_percent[cols_null_percent>0].sort_values(ascending=False)
display(cols_with_null)
```

```python
# если слишком много колонок и цифры сливаются, полезно построить столбчатую диаграмму
cols_with_null.plot(
    kind='bar',
    figsize=(10, 4),
    title='Распределение пропусков в данных'
);
```

```python
# можно создать тепловую карту пропусков, желтым отмечено там где пропуски, но не так удобно
colors = ['blue', 'yellow'] 
fig = plt.figure(figsize=(10, 4))
cols = cols_with_null.index
ax = sns.heatmap(
    sber_data[cols].isnull(),
    cmap=sns.color_palette(colors),
)
```
