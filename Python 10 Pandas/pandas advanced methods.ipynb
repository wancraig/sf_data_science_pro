```python
# Сгруппируйте данные по признаку количества комнат и найдите среднюю цену объектов недвижимости в каждой группе. 
# В качестве ответа запишите количество комнат, для которых средняя цена наибольшая.

# Группировка данных по количеству комнат и вычисление средней цены для каждой группы
mean_price_by_rooms = melb_df.groupby('Rooms')['Price'].mean().reset_index()

# Нахождение количества комнат с наибольшей средней ценой
room_with_max_avg_price = mean_price_by_rooms.loc[mean_price_by_rooms['Price'].idxmax(), 'Rooms']

# Вывод количества комнат с наибольшей средней ценой
print(f"Количество комнат с наибольшей средней ценой: {room_with_max_avg_price}")
```

```python
# Какой регион имеет наименьшее стандартное отклонение по географической широте (Lattitude)? 
# В качестве ответа запишите название этого региона.

# Группировка данных по регионам и вычисление стандартного отклонения широты (Latitude) для каждого региона
std_dev_by_region = melb_df.groupby('Regionname')['Lattitude'].std().reset_index()

# Нахождение региона с наименьшим стандартным отклонением широты
region_with_min_std_dev = std_dev_by_region.loc[std_dev_by_region['Lattitude'].idxmin(), 'Regionname']

# Вывод названия региона с наименьшим стандартным отклонением широты
print(f"Регион с наименьшим стандартным отклонением широты: {region_with_min_std_dev}")
```

```python
# Какая риелторская компания (SellerG) имеет наименьшую общую выручку за период с 1 мая по 1 сентября (вкл) 2017 года? 
# Для ответа на этот вопрос рассчитайте сумму продаж (Price) каждой компании в заданный период.
# Не забудьте перевести даты в формат datetime.

# Преобразование столбца 'Date' в формат datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

# Фильтрация данных за период с 1 мая по 1 сентября 2017 года
filtered_df = melb_df[(melb_df['Date'] >= '2017-05-01') & (melb_df['Date'] <= '2017-09-01')]

# Группировка данных по риелторской компании и вычисление общей выручки (суммы продаж) для каждой компании
total_sales_by_seller = filtered_df.groupby('SellerG')['Price'].sum().reset_index()

# Нахождение компании с наименьшей общей выручкой за заданный период
seller_with_min_sales = total_sales_by_seller.loc[total_sales_by_seller['Price'].idxmin(), 'SellerG']

# Вывод названия компании с наименьшей общей выручкой за период
print(f"Риелторская компания с наименьшей общей выручкой за период: {seller_with_min_sales}")
```

```python
# Создание сводной таблицы с медианной площадью здания по типу объекта недвижимости и количеству комнат
pivot_table = melb_df.pivot_table(values='BuildingArea', index='Type', columns='Rooms', aggfunc='median')

# Нахождение комбинации признаков с наибольшей площадью здания
max_building_area = pivot_table.stack().idxmax()
max_building_area_type, max_building_area_rooms = max_building_area

# Вывод комбинации признаков с наибольшей площадью здания
print(f"Комбинация с наибольшей площадью здания: {max_building_area_type},{max_building_area_rooms}")
```

```python
# Создание сводной таблицы с медианными ценами по риелторскому агентству и типу здания
pivot_table = melb_df.pivot_table(values='Price', index='SellerG', columns='Type', aggfunc='median')

# Нахождение агентства с максимальной медианной ценой для зданий типа 'unit'
max_price_agency = pivot_table['unit'].idxmax()

# Вывод названия агентства с максимальной медианной ценой для зданий типа 'unit'
print(f"Агентство с максимальной медианной ценой для зданий типа 'unit': {max_price_agency}")
```
