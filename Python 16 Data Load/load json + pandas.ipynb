```python
# Импортируем модуль json
import json 
# Импортируем функцию pprint()
from pprint import pprint 
# Импортируем модуль pandas
import pandas as pd 
# Открываем файл и связываем его с объектом "f"
with open('data/recipes.json') as f: 
    # Загружаем содержимое открытого файла в переменную recipes
    recipes = json.load(f) 
# Создаём объект DataFrame из списка recipes
df = pd.DataFrame(recipes) 
# Выводим на экран первые строки полученного DataFrame
display(df.head())
df.info()
```

```python
df2 = pd.DataFrame(recipes)
display(df2)
```

```python
# Полученный вид не является практичным, нужно преобразовать df чтобы каждый ингридиент был представлен и посчитан

# Этап 1: Создание функции для проверки наличия ингредиента
def ingredient_present(ingredient_list, ingredient_to_check):
    """Проверяет, присутствует ли ингредиент в списке ингредиентов."""
    return 1 if ingredient_to_check in ingredient_list else 0

# Этап 2: Создание реестра уникальных ингредиентов и заполнение DataFrame
unique_ingredients = set()
for ingredients_list in df2['ingredients']:
    unique_ingredients.update(ingredients_list)

# Добавление столбца с общим количеством уникальных ингредиентов
df2['total_unique_ingredients'] = df2['ingredients'].apply(lambda x: len(x))
# Запоминаем созданный столбец и удаляем его из текущего положения
total_unique_ingredients_column = df2['total_unique_ingredients']
df2 = df2.drop('total_unique_ingredients', axis=1)

# Вставляем столбец после столбца 'ingredients' (на четвертую позицию, т.к. индексация начинается с 0)
df2.insert(3, 'total_unique_ingredients', total_unique_ingredients_column)

# Этап 3: итерация по реестру и создание новых столбцов в df
for ingredient in unique_ingredients:
    df2[ingredient] = df2['ingredients'].apply(lambda x: ingredient_present(x, ingredient))

display(df2)
```

```python
# Предполагая, что df2 - это ваш DataFrame
all_ingredients = set()

for ingredients_list in df2['ingredients']:
    all_ingredients.update(ingredients_list)

# Количество уникальных ингредиентов
num_unique_ingredients = len(all_ingredients)
print(f"Количество уникальных ингредиентов: {num_unique_ingredients}")
```

```python
# Если мы планируем продолжать работать с DataFrame, созданными на основе данных, которые мы получили в JSON-формате,
# то полезно будет сохранить промежуточный DataFrame в виде CSV-файла. Для выполнения этой операции воспользуемся известной
# нам в Pandas функцией to_csv():
df.to_csv('data/recipes.csv', index = False)

# В качестве основного параметра мы указали имя файла, в котором необходимо сохранить данные. Также мы установили значение
# параметра index как False. Такая настройка позволит нам не сохранять индексы строк в виде отдельного столбца;
# в результате не будут загружаться «лишние» данные при открытии файла при помощи функции read_csv().
```
